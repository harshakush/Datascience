```python
from IPython.display import display, Markdown
a = Image(filename='python-string-indexing.jpg') 
b = Image(filename='Python-String-Slicing-Syntax.png')
c = Image(filename='Python-String-Slicing-Illustration.png')
d = Image(filename='Python-String-Slicing-Negative-Indices.png')
e = Image(filename='Python-String-Slicing-Specifying-Step-Size.png')
string = '<h1>String Indexing and Slicing<h1>'
display(Markdown(string))
display(a)
string = '''<h3>Slicing Syntax<h3>
<h5>If S is a string, the expression S [ start : stop : step ] returns the portion of the string from index <font color="green">start</font> to index <font color="green">stop</font>, at a step size <font color="green">step</font>.<h5>'''
display(Markdown(string))
display(b)
string = """<h5>Basic Example:<h5>
<h6>S = 'ABCDEFGHI'<br>
print(S[2:7]) #CDEFG
<h6>"""
S = 'ABCDEFGHI'
display(Markdown(string))
display(c)
print('*'*20)
print('" S[2:7] =',S[2:7],'"')
print('*'*20)
string = """<h5>Slice with Negative Indices:<h5>
<h6>S = 'ABCDEFGHI'<br>
print(S[-7:-2]) # CDEFG
<h6>"""
display(Markdown(string))
display(d)
S = 'ABCDEFGHI'
print('*'*20)
print('" S[-7:-2] =',S[-7:-2],'"')
print('*'*20)
```

```python
string = '''<h5>Specify Step of the Slicing<h5>
<h6># Return every 2nd item between position 2 to 7<br>
S = 'ABCDEFGHI'<br>
print(S[2:7:2])# CEG
<h6>''' 
display(Markdown(string))
display(e)
S = 'ABCDEFGHI'
print('*'*20)
print('" S[2:7:2] =',S[2:7:2],'"')
print('*'*20)
string='''<h5>Negative Step Size<h5>
# Returns every 2nd item between position 6 to 1 in reverse order<br>
S = 'ABCDEFGHI'<br>
print(S[6:1:-2])    # GEC'''
display(Markdown(string))
S = 'ABCDEFGHI' 
print('*'*20)
print('" S[6:1:-2] = {} "'.format(S[6:1:-2])) 
print('*'*20)
string = '''<h5>Slice with Positive & Negative Indices<h5>
<h6>S = 'ABCDEFGHI'<br>
print(S[2:-5]) # CD<h6>'''
display(Markdown(string))
S = 'ABCDEFGHI' 
print('*'*20)
print('" S[2:-5] =',S[2:-5],'"')
print('*'*20)
string = '''<h5>Slice at Beginning & End<h5>
<h6>Omitting the <font color="green">start</font>  index starts the slice from the index 0. Meaning, S[:stop] is equivalent to S[0:stop]
# Slice first three characters from the string<br>
S = 'ABCDEFGHI'<br>
print(S[:3])    # ABC<h6>'''
display(Markdown(string))
S = 'ABCDEFGHI'
print('*'*20)
print('" S[:3] =',S[:3],'"')
print('*'*20)
string = '''<h6>Whereas, omitting the <font color="green">stop</font>  index extends the slice to the end of the string. Meaning, S[start:] is equivalent to S[start:len(S)]<br>
# Slice last three characters from the string<br>
S = 'ABCDEFGHI'<br>
print(S[6:])    # GHI<h6>'''
display(Markdown(string))
S = 'ABCDEFGHI'
print('*'*20)
print('" S[6:] =',S[6:],'"')
print('*'*20)
string = '''<h5>Reverse a String<h5>
<h6>You can reverse a string by omitting both start and stop indices and specifying a step as -1.<br>
S = 'ABCDEFGHI'<br>
print(S[::-1])    # IHGFEDCBA<h6>'''
display(Markdown(string))
S = 'ABCDEFGHI'
print('*'*25)
print('" S[::-1] =',S[::-1],'"') 
print('*'*25)
```
